For easy understanding am going to scrape multiple jumia pages for amazing wristwatches...let's dive-in it's gonna be alot of fun

```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
    Mens_wrist_watch = []
    for i in range(1,51):
     url = "https://www.jumia.com.ng/mens-formal-watches/?sort=rating&page={i}#catalog-listing"
     response = requests.get(url)
     response = response.content
     soup = BeautifulSoup(response, 'lxml')
     search_tree = soup.find('div', class_ = '-paxs row _no-g _4cl-3cm-shs')
     articles = search_tree.find_all('article', class_ = 'prd _fb col c-prd')
     for article in articles :
       watch_name = article.find('h3', class_ = 'name').text
       watch_price = article.find('div', class_ = 'prc').text
       watch_rating = article.find('div', class_='stars _s').text
       Mens_wrist_watch.append([watch_name, watch_price, watch_rating])
  
```

```python
Mens_wrist_watch
```

```python
Fashionable_Men = pd.DataFrame(Mens_wrist_watch, columns = ['Collection', 'Price', 'Star Ratings'] )
```
